# EDA. Dataset: 'Credit_Card_Fraud_' from OpenML

https://www.openml.org/search?type=data&status=active&id=45955&sort=runs

```python
import pandas as pd
from scipy.io import arff
import matplotlib.pyplot as plt

data, meta = arff.loadarff("../datasets/Credit_Card_Fraud_.arff")

df = pd.DataFrame(data)

for column in df.columns:
    if df[column].dtype == "object":
        df[column] = df[column].str.decode("utf-8")

df.to_csv("../datasets/Credit_Card_Fraud_.csv", index=False)

print(f"rows: {len(df)}")
df.head()
```

### Category of the data

This dataset captures transaction patterns and behaviors that could indicate potential fraud in card transactions. The data is composed of several features designed to reflect the transactional context such as geographical location, transaction medium, and spending behavior relative to the user's history.

### 'repeat_retailer'

Most of the transactions (88.2%) are on the repeated retailers

```python
# df.repeat_retailer

type_values = df.repeat_retailer.value_counts()

plt.pie(
    type_values.values,
    labels=type_values.index,
    autopct=lambda pct: f"{pct:.1f}%" if pct > 1 else "",
    startangle=90,
    wedgeprops={"edgecolor": "black"},
)
plt.title("'repeat_retailer' values")
plt.show()
```
