### EDA. Dataset: 'Credit Card Fraud Detection' from Kaggle

https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../datasets/creditcard.csv')
print(f'rows: {len(df)}')
df.head()
```

### 'Class'

```python
# df.Class

isFraud = df.Class
fraud = isFraud[isFraud == 1]
print(f'Fraud amount: {len(fraud)} ({len(fraud)/len(df.Class)*100:.2}%)')


type_values = df.Class.value_counts()

plt.pie(
    type_values.values,
    labels=type_values.index,
    autopct=lambda pct: f"{pct:.1f}%" if pct > 1 else "",
    startangle=90,
    wedgeprops={'edgecolor': 'black'}
)
plt.title("'Class' values")
plt.show()
```
